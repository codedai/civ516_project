```python
from zipfile import ZipFile
import pandas as pd
import numpy as np
import datetime
```

```python
with ZipFile('thunder_bay_gtfs.zip') as myzip:
    shapes_df = pd.read_csv(myzip.open("shapes.txt"))
    routes_df = pd.read_csv(myzip.open("routes.txt"))
    
    trips_df = pd.read_csv(myzip.open("trips.txt"))
    
    stop_times_df = pd.read_csv(myzip.open("stop_times.txt"))
    
    calendar_dates_df = pd.read_csv(myzip.open("calendar_dates.txt"))
    
calendar_dates_df["date"] = calendar_dates_df['date'].apply(lambda x: datetime.date(int(str(x)[:4]), int(str(x)[4:6]), int(str(x)[6:])))
calendar_dates_df['is_weekday'] = calendar_dates_df['date'].apply(lambda x: x.weekday()<5)
calendar_dates_df[calendar_dates_df['is_weekday']].head()
```

```python
weekday_trips = calendar_dates_df[calendar_dates_df.is_weekday].merge(trips_df, how='left').loc[:, ['trip_id', 'route_id']]
weekday_trips.groupby(['trip_id', 'route_id']).count().to_csv("temp.csv")
weekday_trips = pd.read_csv('temp.csv')
weekday_trip_stop_time = weekday_trips.merge(stop_times_df, how='left')
weekday_trip_stop_time['arrival_time'] = weekday_trip_stop_time.arrival_time.apply(lambda x: int(datetime.datetime.timestamp(datetime.datetime.strptime("25/05/99 " + x, "%d/%m/%y %H:%M:%S"))))
weekday_trip_stop_time
```

```python
# round-trip schedule time for each route
trip_id_list = weekday_trip_stop_time.trip_id.unique()
_route_list = []
_schedule_time_list = []
_start_stop = []
_start_time = []
_end_time = []
for trip_id in trip_id_list:
    _trip_info = weekday_trip_stop_time[weekday_trip_stop_time.trip_id==trip_id]
    _route_list.append(_trip_info.iloc[0, 1])
    _start_time.append(_trip_info.iloc[0, 2])
    _end_time.append(_trip_info.iloc[-1, 2])
    _start_stop.append(_trip_info.iloc[0, 4])
    _sch_time = (max(_trip_info.arrival_time) - min(_trip_info.arrival_time)) / 60
    _schedule_time_list.append(_sch_time)
```

```python
df_3 = pd.DataFrame(np.array([_route_list, trip_id_list, _start_time, _end_time, _schedule_time_list, _start_stop]).T,
             columns=["route_id", "trip_id", "start_time", "end_time", "scheduled_time", "start_stop"])
df_3
```

```python
_route_list = df_3.route_id.unique()
headway_list = []
for route_id in _route_list:
    _route_info = df_3[df_3.route_id==route_id]
    _start_stop = _route_info.start_stop.unique()
    _scheduled_time = _route_info.scheduled_time.unique()
    _max_head = 0
    _max_head_time = None
    _min_head = 24 * 60 * 60
    _min_head_time = None
    for start_stop in _start_stop:
        _route_with_stop = _route_info[_route_info.start_stop==start_stop].copy()
        
        # Find min first, no sorted needed
        _start_time_list = _route_with_stop.start_time.to_numpy()
        _un_sub = np.subtract(_start_time_list[1:], _start_time_list[:-1])
        _sub = sorted(_un_sub)
        for i in _sub:
            if i > 0:
                _tmp_min = i
                break
        if _tmp_min / 60 < _min_head:
            _min_head = _tmp_min / 60
            _min_index = _un_sub.tolist().index(_tmp_min)
            _min_head_time = [_start_time_list[_min_index], _start_time_list[_min_index+1]]
            _min_head_time = [str(datetime.datetime.fromtimestamp(_time).time()) for _time in _min_head_time]
            
        # Find max, need to sort the current list based on start_time
        _s_route_with_stop = _route_with_stop.sort_values(by='start_time')
        
        _start_time_list = _s_route_with_stop.start_time.to_numpy()
        _un_sub = np.subtract(_start_time_list[1:], _start_time_list[:-1])
        _sub = sorted(_un_sub)
        
        _tmp_max = _sub[-1]
        
        if _tmp_max / 60 > _max_head:
            _max_head = _tmp_max / 60
            _max_index = _un_sub.tolist().index(_tmp_max)
            _max_head_time = [_start_time_list[_max_index], _start_time_list[_max_index+1]]
            _max_head_time = [str(datetime.datetime.fromtimestamp(_time).time()) for _time in _max_head_time]

    headway_list.append([route_id, _min_head, _min_head_time, _max_head, _max_head_time, _max_head-_min_head, [str(datetime.datetime.fromtimestamp(min(_route_info.start_time.to_numpy())).time()), str(datetime.datetime.fromtimestamp(max(_route_info.start_time.to_numpy())).time())], (max(_route_info.start_time.to_numpy())-min(_route_info.start_time.to_numpy()))/60, sorted(_scheduled_time)])
```
